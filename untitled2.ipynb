# **–í—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏**

```python
# –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ü—É–∞—Å—Å–æ–Ω–∞
def Puasonchik(lmbda, j):
    if j < 0:
        return 0
    if j == 0:
        return math.exp(-lmbda)
    return (lmbda ** j) * math.exp(-lmbda) / factorial(j)


# –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å "—É—Å–ø–µ—Ö–∞" –¥–ª—è i –∞–∫—Ç–∏–≤–Ω—ã—Ö –∞–±–æ–Ω–µ–Ω—Ç–æ–≤
def uspex(i):
    if i == 1:
        return 1
    if i <= 0:
        return 0
    return (1 - 1 / i) ** (i - 1)
```

# **–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –ø–µ—Ä–µ—Ö–æ–¥–æ–≤**

```python
def from0toj(lmbda, j):
    return Puasonchik(lmbda, j)
"""
–≠—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è –≤—ã—á–∏—Å–ª—è–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–µ—Ä–µ—Ö–æ–¥–∞ –∏–∑ —Å–æ—Å—Ç–æ—è–Ω–∏—è "0" –≤ —Å–æ—Å—Ç–æ—è–Ω–∏–µ j,
—á—Ç–æ —è–≤–ª—è–µ—Ç—Å—è –ø–µ—Ä–µ—Ö–æ–¥–æ–º –æ—Ç –Ω—É–ª–µ–≤–æ–≥–æ —á–∏—Å–ª–∞ –∞–∫—Ç–∏–≤–Ω—ã—Ö –∞–±–æ–Ω–µ–Ω—Ç–æ–≤ –∫ j –∞–∫—Ç–∏–≤–Ω—ã–º –∞–±–æ–Ω–µ–Ω—Ç–∞–º.
"""

def i_to_prev_i(lmbda, i):
    if i < 1:
        return 0
    return uspex(i) * Puasonchik(lmbda, 0)

"""
–≠—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–µ—Ä–µ—Ö–æ–¥–∞ –∏–∑ —Å–æ—Å—Ç–æ—è–Ω–∏—è i –≤ —Å–æ—Å—Ç–æ—è–Ω–∏–µ i,
 —Ç–æ –µ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–∏–≤–Ω—ã—Ö –∞–±–æ–Ω–µ–Ω—Ç–æ–≤ –Ω–µ –∏–∑–º–µ–Ω–∏—Ç—Å—è.
"""

def stayi(lmbda, i):
    return (1 - uspex(i)) * Puasonchik(lmbda, 0) + uspex(i) * Puasonchik(lmbda, 1)
"""
–≠—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è –≤—ã—á–∏—Å–ª—è–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å–∏—Å—Ç–µ–º–∞ –æ—Å—Ç–∞–Ω–µ—Ç—Å—è –≤ —Å–æ—Å—Ç–æ—è–Ω–∏–∏ i
–í —Ç–µ–æ—Ä–∏–∏ –æ–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è –∫–∞–∫ –¥–≤–∞ –≤–æ–∑–º–æ–∂–Ω—ã—Ö —Å—Ü–µ–Ω–∞—Ä–∏—è:
–û–¥–∏–Ω –∞–±–æ–Ω–µ–Ω—Ç –ø–µ—Ä–µ–¥–∞–µ—Ç —Å–æ–æ–±—â–µ–Ω–∏–µ, –∏ –Ω–æ–≤—ã–π –∞–±–æ–Ω–µ–Ω—Ç –Ω–µ –ø–æ—è–≤–ª—è–µ—Ç—Å—è: –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –¥–ª—è —ç—Ç–æ–≥–æ —Å–æ–±—ã—Ç–∏—è ‚Äî
–ù–æ–≤—ã–π –∞–±–æ–Ω–µ–Ω—Ç –ø–æ—è–≤–ª—è–µ—Ç—Å—è, –∏ –Ω–∏–∫—Ç–æ –Ω–µ –ø–µ—Ä–µ–¥–∞–µ—Ç —Å–æ–æ–±—â–µ–Ω–∏–µ: –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –¥–ª—è —ç—Ç–æ–≥–æ —Å–æ–±—ã—Ç–∏—è ‚Äî

–í –∫–æ–¥–µ —ç—Ç–∏ –¥–≤–∞ —Å–æ–±—ã—Ç–∏—è –æ–±—ä–µ–¥–∏–Ω–µ–Ω—ã —Å –ø–æ–º–æ—â—å—é –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ —Å–ª–æ–∂–µ–Ω–∏—è. –≠—Ç–æ —Ç–∞–∫–∂–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —Ç–µ–æ—Ä–∏–∏, –≥–¥–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ—Å—Ç–∞–≤–∞—Ç—å—Å—è –≤ —Å–æ—Å—Ç–æ—è–Ω–∏–∏ i
i ‚Äî —ç—Ç–æ —Å—É–º–º–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π —ç—Ç–∏—Ö –¥–≤—É—Ö —Å—Ü–µ–Ω–∞—Ä–∏–µ–≤.
"""

def i_to_j(lmbda, i, j):
    if j < i:
        return 0
    return uspex(i) * Puasonchik(lmbda, j - i + 1) + (1 - uspex(i)) * Puasonchik(lmbda, j - i)

"""
–≠—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è –≤—ã—á–∏—Å–ª—è–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–µ—Ä–µ—Ö–æ–¥–∞ –∏–∑ —Å–æ—Å—Ç–æ—è–Ω–∏—è i –≤ —Å–æ—Å—Ç–æ—è–Ω–∏–µ j,
—Ç–æ –µ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å–∏—Å—Ç–µ–º–∞ –∏–∑ i –∞–±–æ–Ω–µ–Ω—Ç–æ–≤ –ø–µ—Ä–µ–π–¥–µ—Ç –≤ j –∞–±–æ–Ω–µ–Ω—Ç–æ–≤.

–í —Ç–µ–æ—Ä–∏–∏ —ç—Ç–æ—Ç –ø–µ—Ä–µ—Ö–æ–¥ –º–æ–∂–µ—Ç –±—ã—Ç—å:
1)–û–¥–∏–Ω –∞–±–æ–Ω–µ–Ω—Ç –ø–µ—Ä–µ–¥–∞–ª —Å–æ–æ–±—â–µ–Ω–∏–µ, –∞ –ø–æ—è–≤–∏–ª–æ—Å—å j‚àíi+1 –Ω–æ–≤—ã—Ö –∞–±–æ–Ω–µ–Ω—Ç–æ–≤.

2)–ù–∏ –æ–¥–∏–Ω –∞–±–æ–Ω–µ–Ω—Ç –Ω–µ –ø–µ—Ä–µ–¥–∞–ª —Å–æ–æ–±—â–µ–Ω–∏–µ, –Ω–æ –ø–æ—è–≤–∏–ª–æ—Å—å j‚àíi –Ω–æ–≤—ã—Ö –∞–±–æ–Ω–µ–Ω—Ç–æ–≤.
"""
```

# **–°–∏—Å—Ç–µ–º–∞ –ª–∏–Ω–µ–π–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π**

–¶–µ–ª—å —Ñ—É–Ω–∫—Ü–∏–∏ system(lmbda, N) ‚Äî –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –∏ —Ä–µ—à–∏—Ç—å —Å–∏—Å—Ç–µ–º—É –ª–∏–Ω–µ–π–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å–æ—Å—Ç–æ—è–Ω–∏–π –ú–∞—Ä–∫–æ–≤—Å–∫–æ–π —Ü–µ–ø–∏. –ó–¥–µ—Å—å:
lmbda ‚Äî –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å –ø–æ—Ç–æ–∫–∞ —Å–æ–æ–±—â–µ–Ω–∏–π (–ø–∞—Ä–∞–º–µ—Ç—Ä —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞).
N ‚Äî –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–±–æ–Ω–µ–Ω—Ç–æ–≤ –≤ —Å–∏—Å—Ç–µ–º–µ (—Ä–∞–∑–º–µ—Ä —Å–∏—Å—Ç–µ–º—ã).

transition_matrix ‚Äî —ç—Ç–æ –º–∞—Ç—Ä–∏—Ü–∞ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏
(N+1)√ó(N+1), –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç —Ö—Ä–∞–Ω–∏—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –ø–µ—Ä–µ—Ö–æ–¥–∞ –∏–∑ –æ–¥–Ω–æ–≥–æ —Å–æ—Å—Ç–æ—è–Ω–∏—è –≤ –¥—Ä—É–≥–æ–µ.

–î–∞–ª–µ–µ, –≤ —Ü–∏–∫–ª–µ –∑–∞–ø–æ–ª–Ω—è–µ—Ç—Å—è —ç—Ç–∞ –º–∞—Ç—Ä–∏—Ü–∞. –î–ª—è –∫–∞–∂–¥–æ–≥–æ —Å–æ—Å—Ç–æ—è–Ω–∏—è i –∏ j –∑–∞–¥–∞—é—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–µ –ø—Ä–∞–≤–∏–ª–∞:

```python
# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Å–∏—Å—Ç–µ–º—ã –ª–∏–Ω–µ–π–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π
def system(lmbda, N):
    transition_matrix = np.zeros((N + 1, N + 1))
    for i in range(N + 1):
        for j in range(N + 1):
            if i == 0: #—Å–æ—Å—Ç–æ—è–Ω–∏–µ "–Ω—É–ª–µ–≤–æ–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏", –∫–æ–≥–¥–∞ –Ω–µ—Ç –∞–±–æ–Ω–µ–Ω—Ç–æ–≤ —Å —Å–æ–æ–±—â–µ–Ω–∏—è–º–∏
                transition_matrix[i, j] = from0toj(lmbda, j)
            elif i == j:#—Å–æ—Å—Ç–æ—è–Ω–∏–µ –Ω–µ –º–µ–Ω—è–µ—Ç—Å—è
                transition_matrix[i, j] = stayi(lmbda, i)
            elif i < j:#–ø–µ—Ä–µ—Ö–æ–¥ –∏–∑ –º–µ–Ω—å—à–µ–≥–æ —Å–æ—Å—Ç–æ—è–Ω–∏—è –≤ –±–æ–ª—å—à–µ–µ
                transition_matrix[i, j] = i_to_j(lmbda, i, j)
            elif i == j + 1:#–ø–µ—Ä–µ—Ö–æ–¥ –≤ —Å–æ—Å—Ç–æ—è–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –Ω–∞ 1 –º–µ–Ω—å—à–µ —Ç–µ–∫—É—â–µ–≥–æ
                transition_matrix[i, j] = i_to_prev_i(lmbda, i)
            else: #–ï—Å–ª–∏ –Ω–∏–∫–∞–∫–æ–µ –∏–∑ —ç—Ç–∏—Ö —É—Å–ª–æ–≤–∏–π –Ω–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è, –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–µ—Ä–µ—Ö–æ–¥–∞ —Ä–∞–≤–Ω–∞ 0
                transition_matrix[i, j] = 0

#–ü–æ—Å–ª–µ —Ç–æ–≥–æ –∫–∞–∫ –º–∞—Ç—Ä–∏—Ü–∞ –ø–µ—Ä–µ—Ö–æ–¥–æ–≤ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∞, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Ä–µ—à–∏—Ç—å —Å–∏—Å—Ç–µ–º—É –ª–∏–Ω–µ–π–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π, —á—Ç–æ–±—ã –Ω–∞–π—Ç–∏ —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–æ—Å—Ç–æ—è–Ω–∏–π.

#transition_matrix.T ‚Äî —ç—Ç–æ —Ç—Ä–∞–Ω—Å–ø–æ–Ω–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ –ø–µ—Ä–µ—Ö–æ–¥–æ–≤.
#np.eye(N + 1) ‚Äî —ç—Ç–æ –µ–¥–∏–Ω–∏—á–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ —Ä–∞–∑–º–µ—Ä–æ–º (N+1)√ó(N+1).
#matrix ‚Äî —ç—Ç–æ –Ω–æ–≤–∞—è –º–∞—Ç—Ä–∏—Ü–∞, –∫–æ—Ç–æ—Ä–∞—è –ø–æ–ª—É—á–∞–µ—Ç—Å—è –ø—É—Ç—ë–º –≤—ã—á–∏—Ç–∞–Ω–∏—è –µ–¥–∏–Ω–∏—á–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã –∏–∑ —Ç—Ä–∞–Ω—Å–ø–æ–Ω–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã –ø–µ—Ä–µ—Ö–æ–¥–æ–≤ –∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –ø–æ—Å–ª–µ–¥–Ω–µ–π —Å—Ç—Ä–æ–∫–∏, —Å–æ—Å—Ç–æ—è—â–µ–π –∏–∑ –µ–¥–∏–Ω–∏—Ü.

    # –†–µ—à–µ–Ω–∏–µ —Å–∏—Å—Ç–µ–º—ã —É—Ä–∞–≤–Ω–µ–Ω–∏–π
    matrix = np.vstack([transition_matrix.T - np.eye(N + 1), np.ones(N + 1)])
    coeffs = np.zeros(N + 2)#–≤–µ–∫—Ç–æ—Ä –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ coeffs
    coeffs[-1] = 1#–≠—Ç–æ—Ç –≤–µ–∫—Ç–æ—Ä —Å–æ–¥–µ—Ä–∂–∏—Ç 0 –≤ –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–µ –ø–æ–∑–∏—Ü–∏–π, –∑–∞ –∏—Å–∫–ª—é—á–µ–Ω–∏–µ–º –ø–æ—Å–ª–µ–¥–Ω–µ–π, –∫–æ—Ç–æ—Ä–∞—è —Ä–∞–≤–Ω–∞ 1.
    pi = np.linalg.lstsq(matrix, coeffs, rcond=None)[0]#np.linalg.lstsq ‚Äî —ç—Ç–æ —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–µ—à–µ–Ω–∏—è —Å–∏—Å—Ç–µ–º—ã –ª–∏–Ω–µ–π–Ω—ã—Ö —É—Ä–∞–≤–Ω–µ–Ω–∏–π –º–µ—Ç–æ–¥–æ–º –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤.
    return pi#–§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –≤–µ–∫—Ç–æ—Ä œÄ, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç —Å–æ–±–æ–π —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ú–∞—Ä–∫–æ–≤—Å–∫–æ–π —Ü–µ–ø–∏:
```

# **–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–±–æ–Ω–µ–Ω—Ç–æ–≤**

 –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∞–±–æ–Ω–µ–Ω—Ç–æ–≤, –Ω–∞—Ö–æ–¥—è—â–∏—Ö—Å—è –≤ —Å–∏—Å—Ç–µ–º–µ, –¥–ª—è —Ä–∞–∑–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç–∏ –≤—Ö–æ–¥–Ω–æ–≥–æ –ø–æ—Ç–æ–∫–∞

K, –∞ —Ç–∞–∫–∂–µ –≥–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Å –∏–º–∏—Ç–∞—Ü–∏–æ–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏.

```python
# –°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–±–æ–Ω–µ–Ω—Ç–æ–≤
def mean_abonents(lambdas, K):
    mean = []
    for lmbda in lambdas:
        stat = system(lmbda, K)
        average_users = np.dot(stat, np.arange(K + 1))
        mean.append(average_users)
    return mean

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
final_result = []
for K in Numbers:
    mean = mean_abonents(lambda_, K)
    final_result.append(mean)

# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –∏–º–∏—Ç–∞—Ü–∏–æ–Ω–Ω—ã—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –¥–ª—è –ø—Ä–∏–º–µ—Ä–∞
result_list_ns = [np.mean(results) for results in zip(*final_result)]
```

# **–ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞**

```python
def plot_average_AB(lambda_values, average_results, K_values, simulation_results):
    """
    –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∞–±–æ–Ω–µ–Ω—Ç–æ–≤ –æ—Ç –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç–∏ Œª.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - lambda_values: —Å–ø–∏—Å–æ–∫ –∑–Ω–∞—á–µ–Ω–∏–π Œª.
    - average_results: —Å–ø–∏—Å–æ–∫ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è —Ä–∞–∑–Ω—ã—Ö K.
    - K_values: —Å–ø–∏—Å–æ–∫ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∞–±–æ–Ω–µ–Ω—Ç–æ–≤ (K).
    - simulation_results: —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –∏–º–∏—Ç–∞—Ü–∏–æ–Ω–Ω–æ–≥–æ –º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏—è.
    """
    plt.figure(figsize=(12, 7))

    # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ —Ü–≤–µ—Ç–æ–≤–æ–π –ø–∞–ª–∏—Ç—Ä—ã –∏ –º–∞—Ä–∫–µ—Ä–æ–≤
    colors = plt.cm.viridis(np.linspace(0, 1, len(K_values)))
    markers = ['o', 's', 'D', '^', 'v']

    # –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è K
    for idx, (K, avg_result) in enumerate(zip(K_values, average_results)):
        plt.plot(lambda_values, avg_result,
                 label=f"–ú–∞–∫—Å. –∞–±–æ–Ω–µ–Ω—Ç–æ–≤ = {K}",
                 marker=markers[idx % len(markers)],
                 linestyle='-',
                 color=colors[idx],
                 markersize=8,
                 linewidth=2)

    # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è –∏–º–∏—Ç–∞—Ü–∏–æ–Ω–Ω–æ–≥–æ –º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏—è
    plt.plot(lambda_values, simulation_results,
             label="–ò–º–∏—Ç–∞—Ü–∏–æ–Ω–Ω–æ–µ –º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏–µ –¥–ª—è N = 100000",
             marker='o',
             linestyle='-',
             color='purple',
             markersize=8,
             linewidth=3)

    # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –ø–æ–¥–ø–∏—Å–∏ –æ—Å–µ–π –∏ –∑–∞–≥–æ–ª–æ–≤–∫–∞
    plt.xlabel("Œª", fontsize=14, fontweight='bold')
    plt.ylabel("–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–±–æ–Ω–µ–Ω—Ç–æ–≤", fontsize=14, fontweight='bold')
    plt.yscale("log")  # –õ–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–∞—è —à–∫–∞–ª–∞ –æ—Å–∏ Y
    plt.title("–ì—Ä–∞—Ñ–∏–∫ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–∏—Å–ª–∞ –∞–±–æ–Ω–µ–Ω—Ç–æ–≤ –æ—Ç –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç–∏ Œª", fontsize=16, fontweight='bold')

    # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –ª–µ–≥–µ–Ω–¥—ã –∏ —Å–µ—Ç–∫–∏
    plt.legend(fontsize=12)
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)

    # –û–ø—Ç–∏–º–∏–∑–∞—Ü–∏—è —Ä–∞–∑–º–µ—â–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
    plt.tight_layout()
    plt.show()

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
plot_average_AB(lambda_, final_result, Numbers, result_list_ns)
```
